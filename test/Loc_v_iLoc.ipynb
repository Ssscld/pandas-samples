```python
# importing pandas and numpy
import pandas as pd
import numpy as np

# crete a sample dataframe
data = pd.DataFrame({
    'age' :     [ 10, 22, 13, 21, 12, 11, 17],
    'section' : [ 'A', 'B', 'C', 'B', 'B', 'A', 'A'],
    'city' :    [ 'Gurgaon', 'Delhi', 'Mumbai', 'Delhi', 'Mumbai', 'Delhi', 'Mumbai'],
    'gender' :  [ 'M', 'F', 'F', 'M', 'M', 'M', 'F'],
    'favourite_color' : [ 'red', np.NAN, 'yellow', np.NAN, 'black', 'green', 'red']
})

# view the data
data
```

```python
#Condition
data.loc[data['age'] >= 15]

## Note and iLoc use '[]' instead of '()'
## This works as well (ie no loc, direct dereference) : data[data['age'] >= 15]
```

```python
# multiple conditions = note the parens around conditions, Note single &;  use | for OR and ~ for NOT
data.loc[(data['age'] >=12) & (data['gender']=='M')]
```

```python
data.loc[1:3]   #Data frame needs to be sorted on index. Else will select rows 1 and 3
## note the upper bound is included in the display
```

```python
# select few columns with a condition
data.loc[(data.age >= 12), ['city', 'gender']]
```

```python
# update a column with condition
data.loc[(data.age >= 12), ['section']] = 'M'
data

## note the assignment is done 'inplace' - no specific flag needed like other functions. 
```

```python

# update multiple columns with condition
data.loc[(data.age >= 20), ['section', 'city']] = ['S','Pune']
data
```

```python
# select rows with indexes
data.iloc[[0,2]]
```

```python
# select rows with particular indexes and particular columns
data.iloc[[0,2],[1,3]]

## first series of brackets is for row list (0,2), 2nd series is for column list (1,3)
```

```python

# select a range of rows and columns
data.iloc[1:3,2:4]
```
